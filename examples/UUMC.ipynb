```python
M = 100 #100 and 10
Datasets = {k: CausalModel.gen_dataset(10, 500, M, coef_args={'style':s}) for k, s in GEN_types.items()}
```

```python
algs = [PC(), PC('stable'), Notears(), GES(), boss]
alg_names = ["PC", "PC-stable", "NOTEARS", "GES", "boss"]
```

```python
trap = io.StringIO()

import logging, sys
logging.disable(sys.maxsize)

Metric_list = []
for i, alg in enumerate(algs):
    a_name = alg_names[i]
    Metrics = np.ones((len(Datasets.keys()),M))
    for d, (k, D) in enumerate(Datasets.items()):
        for g, G in enumerate(D):
            Dag = G.A
            _progress_message('{}% completed'.format(round((g+d*M)*100/(M*len(GEN_types)))))
            with warnings.catch_warnings():
                warnings.simplefilter("ignore")
                if a_name == "boss":
                    with redirect_stdout(trap):
                        BG = alg(G.data.data.T).graph
                        BG_oriented = ((BG<0)*(BG.T>0))*1
                        BG_bod = (BG<0)*1
                        m_o = MetricsDAG(BG_oriented, G.A!=0).metrics['F1']
                        m_bod = MetricsDAG(BG_bod, G.A!=0).metrics['F1']
                        Metrics[d,g] = (m_o + m_bod)/2 #half credit for unoriented edges
                else:
                    alg.learn(G.data.data.T)
                    Metrics[d,g] = MetricsDAG(alg.causal_matrix, Dag).metrics['F1']
            if np.isnan(Metrics[d,:]).any():
                nm = np.nanmean(Metrics[d,:])
                Metrics[d,:][np.isnan(Metrics[d,:])] = nm
    Metric_list+=[Metrics]
    _clear_progress_message("{} down, {} to go!".format(len(Metric_list), len(algs)-len(Metric_list)))
```

```python
new_list = [np.concatenate([Metric_list[alg][[gen_type],:] for alg in range(len(algs))], axis=0) for gen_type in range(len(GEN_types))]
```

```python
colors1 = ["blue", "purple", "orange", "green", "red"]
colors2 = ["skyblue", "plum", "navajowhite", "lightgreen", "peachpuff"]
w = .15
for i, Metrics in enumerate(new_list):
    SortabilityPlotting.plot_box(Metrics.T, np.arange(len(alg_names))+w*(i-2), colors1[i], colors2[i], list(GEN_types.keys())[i], widths=w)
plt.ylim([0,1])
ax = plt.gca()
ax.set_xticks(np.arange(len(alg_names)))
ax.set_xticklabels(alg_names)#, rotation='vertical', fontsize=18)
plt.legend(loc=(.5,0.7))
plt.title("F1 scores for Causal Discovery Algorithms on Generated Benchmark Data")
```
